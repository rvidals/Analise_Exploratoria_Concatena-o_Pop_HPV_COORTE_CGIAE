# Código para concatenar vários csv e ordenar a partir do Sexo e o código do Município


```python
#Importando a biblioteca pandas
import pandas as pd

#Importar os arquivos csv e salvando na variável pop
from google.colab import files
pop = files.upload()
```

```python
#Caminho onde serão salvo os arquivos
path = '/content/'

#Caso obte por usar 
#path2 = '/content/PopResidente_SUS/'

li_M = [] # os valores são atribuidos à uma lista
li_F = [] # os valores são atribuidos à uma lista
li_IBGE = [] # o Código do IBGE e nome do município será atribuido a esta lista

# o método key foi usado para chamar os  nomes dos arquivos presentes no google colab
for name in pop.keys():

  #Essa parte é para pegar a faixa etária e renomear a coluna sexo
  df_geral  = pd.read_csv(path+name, encoding='latin1')
  
  sexo = df_geral.iat[1,0]
  sexo = sexo.split(':')[1].replace(' ','')
  
  fEtaria = df_geral.iat[2,0]
  fEtaria = fEtaria.split(':')[1].replace('  ',' ').replace(' ','_')
  
  

  #Essa parte é para pegar o csv que gerará a nava tabela
  # o csv vem com informações além das tabulaeres 
  # Para fazer essa limpeza foi utilizados as flegues skiprows, skipfooter e header
  
  df = pd.read_csv(path+name,
                  encoding='latin1',
                  sep=';',
                  skiprows=5,
                  skipfooter=5,
                  engine='python'
                 )
  df2 = df.rename(columns={"População_residente": sexo+fEtaria})
  
  #Lista onde será salvo as colunas da tabela para concatenar 
  
  if sexo == 'Masculino':
    li_M.append(df2["Masculino"+fEtaria]) # atribuindo as colunas renomeadas à lista
    print(f"{name} - {fEtaria} : {sexo}") #visualizar a execução da extração
  else:
    li_F.append(df2["Feminino"+fEtaria]) # atribuindo as colunas renomeadas à lista
    print(f"{name} - {fEtaria} : {sexo}") #visualizar a execução da extração

li_IBGE.append(df["Município"].str.split(' ',
                                         1,
                                         expand=True).rename(columns = {0: 'COD_MUN',
                                                                       1:'NM_MUN'})) 
```

```python
#Estou inserindo os dados do ibge na lista masculina
li_IBGE_M = li_IBGE + li_M 
#Estou inserindo os dados do ibge na lista masculina
li_IBGE_F = li_IBGE + li_F

#Concatenando os valores em um DataFrame
concat_M = pd.concat(li_IBGE_M , axis=1) # concatenando pelo eixo y = 1
concat_F = pd.concat(li_IBGE_F, axis=1) #Vizualização 

#Inserção da coluna Sexo: M e F
concat_M.insert(2,'Sexo','M')
concat_F.insert(2,'Sexo','F')
```

```python
#Renomeando as colunas 

"""Foi feito um print na estrutura de renomeação utilizando os métodos
.columns.values"""

concat_M = concat_M.rename(columns={'COD_MUN': 'COD_MUN',
'NM_MUN': 'NM_MUN',
'Sexo': 'Sexo',
'Masculino_9_anos': '9_anos',
'Masculino_10_anos': '10_anos',
'Masculino_11_anos': '11_anos',
'Masculino_12_anos': '12_anos',
'Masculino_13_anos': '13_anos',
'Masculino_14_anos': '14_anos',
'Masculino_15_anos': '15_anos'})

concat_F = concat_F.rename(columns={'COD_MUN': 'COD_MUN',
'NM_MUN': 'NM_MUN',
'Sexo': 'Sexo',
'Feminino_9_anos': '9_anos',
'Feminino_10_anos': '10_anos',
'Feminino_11_anos': '11_anos',
'Feminino_12_anos': '12_anos',
'Feminino_13_anos': '13_anos',
'Feminino_14_anos': '14_anos',
'Feminino_15_anos': '15_anos'})

#Concatenando as colunas Masculino e Feminino
popConcat = pd.concat([concat_M,concat_F])

#Ordenando as colunas. 
popConcat = popConcat[['COD_MUN','NM_MUN','Sexo',
                       '9_anos','10_anos','11_anos',
                       '12_anos','13_anos','14_anos',
                       '15_anos']]

```

```python
#Ordenando as linhas a partir das colunas COD_MUN e Sexo e resetando o index
popResid = popConcat.sort_values(['COD_MUN','Sexo']).reset_index(drop=True)

#Salvando o arquivo final em XLSX
popResid.to_excel('Populacao_9a15anos_HPV_CGIAE.xlsx',
                  sheet_name='Pop_HPV',
                  index=False)

#Visualizando o resultado final
popResid
```
